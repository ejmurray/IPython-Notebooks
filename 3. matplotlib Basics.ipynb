##Plot y = cos(x)

```python
x = np.arange(0,10, 1/30)
y = np.cos(x)
```

```python
plt.plot(x,y)
```

##Saving Figures
**Visualization f(x), done by drawing the curve y = f(x) in xy coordinate system**

```python
def f(x):
    return x**2*exp(-x**2)
```

```python
x = linspace(0, 3, 31)
y = f(x)

fig1 = plt.gcf() #Get a reference to the current figure; fig1 refrences the object (image) in memory

plot(x,y)
```

```python
#We can save hardcopy in PDF, PNG or another image format
fig1.savefig('tmp.png')
```

```python
from IPython.display import Image

Image("tmp.png")
```
